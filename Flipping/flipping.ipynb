```python
import cv2
import tkinter as tk
from tkinter import filedialog
from tkinter.filedialog import askopenfile
from PIL import Image, ImageTk
import skimage
import numpy as np

window = tk.Tk()
window.geometry("1000x350")
window.columnconfigure(0, weight=4) 
window.columnconfigure(1, weight=4)
window.columnconfigure(2, weight=4)
window.columnconfigure(3, weight=4)
window.title('Image Flipping')

label_title = tk.Label(window,text='              Image',font=('times', 16))
label_title.grid(row=0, column=0, pady=(20, 10), padx=(20, 0), sticky="w")

button_upload = tk.Button(window, text='Choose File', width=27, command=lambda:upload_file()) 
button_upload.grid(row=1, column=0, pady=(0,20), padx=(20, 0), sticky="w")

label_horizontal = tk.Label(window,text='           Horizontal',font=('times', 16))
label_horizontal.grid(row=0, column=1, rowspan=1, pady=(20, 10), padx=(20, 0), sticky="w")

label_vertikal = tk.Label(window,text='        Vertikal',font=('times', 16))
label_vertikal.grid(row=0, column=2, rowspan=1, pady=(20, 10), padx=(20, 0), sticky="w")

label_kombinasi = tk.Label(window,text='        Kombinasi',font=('times', 16))
label_kombinasi.grid(row=0, column=3, rowspan=1, pady=(20, 10), padx=(20, 0), sticky="w")

def upload_file():
    global label_image_upload
    global label_image_horizontal
    global label_image_vertikal
    global label_image_kombinasi

    try:
        label_image_upload.grid_remove()
        label_image_horizontal.grid_remove()
        label_image_vertikal.grid_remove()
        label_image_kombinasi.grid_remove()
    except:
        print("")
    finally:  
        #Image Upload
        image_type = [('Image File', ['*.jpg', '*.jpeg', '*.png'])]
        image_name = filedialog.askopenfilename(filetypes=image_type)
        image_upload_open = Image.open(image_name)
        image_upload_resize = image_upload_open.resize((197, 197), Image.Resampling.LANCZOS)
        image_upload = ImageTk.PhotoImage(image_upload_resize)
        label_image_upload = tk.Label(window, image=image_upload)
        label_image_upload.image = image_upload
        label_image_upload.grid(row=2, column=0, pady=(0,20), padx=(20, 0), sticky="w")

        #Image Property
        image = cv2.imread(image_name, cv2.IMREAD_UNCHANGED)
        w = image.shape[1]
        h = image.shape[0]
        mode = {'1':1, 'L':8, 'P':8, 'RGB':24, 'RGBA':32, 'CMYK':32, 'YCbCr':24, 'I':32, 'F':32}
        image_depth = mode[image_upload_open.mode]

        #Image Horizontal
        image_horizontal = skimage.io.imread(fname=image_name)
        image_horizontal_temp = skimage.io.imread(fname=image_name)
        if image_depth == 8 or image_depth == 24:
            for y in range(0, h):
                for x in range(0, w):
                    x_ = w - 1 - x
                    image_horizontal[y, x] = image_horizontal_temp[y, x_]
        image_horizontal = Image.fromarray(image_horizontal)
        image_horizontal_resize = image_horizontal.resize((197, 197), Image.Resampling.LANCZOS)
        image_horizontal = ImageTk.PhotoImage(image_horizontal_resize)
        label_image_horizontal = tk.Label(window, image=image_horizontal)
        label_image_horizontal.image = image_horizontal
        label_image_horizontal.grid(row=2, column=1, pady=(0,20), padx=(20, 0), sticky="w")

        #Image Vertikal
        image_vertikal = skimage.io.imread(fname=image_name)
        image_vertikal_temp = skimage.io.imread(fname=image_name)
        if image_depth == 8 or image_depth == 24:
            for y in range(0, h):
                for x in range(0, w):
                    y_ = h - 1 - y
                    image_vertikal[y, x] = image_vertikal_temp[y_, x]
        image_vertikal = Image.fromarray(image_vertikal)
        image_vertikal_resize = image_vertikal.resize((197, 197), Image.Resampling.LANCZOS)
        image_vertikal = ImageTk.PhotoImage(image_vertikal_resize)
        label_image_vertikal = tk.Label(window, image=image_vertikal)
        label_image_vertikal.image = image_vertikal
        label_image_vertikal.grid(row=2, column=2, pady=(0,20), padx=(20, 0), sticky="w")
        
        #Image kombinasi
        image_kombinasi = skimage.io.imread(fname=image_name)
        image_kombinasi_temp = skimage.io.imread(fname=image_name)
        if image_depth == 8 or image_depth == 24:
            for y in range(0, h):
                for x in range(0, w):
                    x_ = w - 1 - x
                    image_kombinasi[y, x] = image_kombinasi_temp[y, x_]
            image_kombinasi_temp = np.array(image_kombinasi, copy=True)
            for y in range(0, h):
                for x in range(0, w):
                    y_ = h - 1 - y
                    image_kombinasi[y, x] = image_kombinasi_temp[y_, x]
        image_kombinasi = Image.fromarray(image_kombinasi)
        image_kombinasi_resize = image_kombinasi.resize((197, 197), Image.Resampling.LANCZOS)
        image_kombinasi = ImageTk.PhotoImage(image_kombinasi_resize)
        label_image_kombinasi = tk.Label(window, image=image_kombinasi)
        label_image_kombinasi.image = image_kombinasi
        label_image_kombinasi.grid(row=2, column=3, pady=(0,20), padx=(20, 0), sticky="w")
        
        

window.mainloop()
```
