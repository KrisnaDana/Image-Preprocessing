```python
import cv2
import tkinter as tk
from tkinter import filedialog
from tkinter.filedialog import askopenfile
from PIL import Image, ImageTk
import skimage

window = tk.Tk()
window.geometry("750x350")
window.columnconfigure(0, weight=4) 
window.columnconfigure(1, weight=4)
window.columnconfigure(2, weight=4)
window.title('Image')

label_title = tk.Label(window,text='              Image',font=('times', 16))
label_title.grid(row=0, column=0, pady=(20, 10), padx=(20, 0), sticky="w")

button_upload = tk.Button(window, text='Choose File', width=27, command=lambda:upload_file()) 
button_upload.grid(row=1, column=0, pady=(0,20), padx=(20, 0), sticky="w")

label_negasi = tk.Label(window,text='              Negasi',font=('times', 16))
label_negasi.grid(row=0, column=1, rowspan=1, pady=(20, 10), padx=(20, 0), sticky="w")

label_threshold = tk.Label(window,text='            Threshold',font=('times', 16))
label_threshold.grid(row=0, column=2, rowspan=1, pady=(20, 10), padx=(20, 0), sticky="w")

def upload_file():
    global label_image_upload
    global label_image_negasi
    global label_image_threshold

    try:
        label_image_upload.grid_remove()
        label_image_negasi.grid_remove()
        label_image_threshold.grid_remove()
    except:
        print("")
    finally:  
        #Image Upload
        image_type = [('Image File', ['*.jpg', '*.jpeg', '*.png'])]
        image_name = filedialog.askopenfilename(filetypes=image_type)
        image_upload_open = Image.open(image_name)
        image_upload_resize = image_upload_open.resize((197, 197), Image.Resampling.LANCZOS)
        image_upload = ImageTk.PhotoImage(image_upload_resize)
        label_image_upload = tk.Label(window, image=image_upload)
        label_image_upload.image = image_upload
        label_image_upload.grid(row=2, column=0, pady=(0,20), padx=(20, 0), sticky="w")

        #Image Property
        image = cv2.imread(image_name, cv2.IMREAD_UNCHANGED)
        image_resolution_x = image.shape[1]
        image_resolution_y = image.shape[0]
        mode = {'1':1, 'L':8, 'P':8, 'RGB':24, 'RGBA':32, 'CMYK':32, 'YCbCr':24, 'I':32, 'F':32}
        image_depth = mode[image_upload_open.mode]

        #Image Negasi
        image_negasi = skimage.io.imread(fname=image_name)
        if image_depth == 8:
            for i in range(0, image_resolution_y):
                for j in range(0, image_resolution_x):
                    image_negasi[i, j] = 255 - image_negasi[i, j]
        elif image_depth == 24:
            for i in range(0, image_resolution_y):
                for j in range(0, image_resolution_x):
                    image_negasi[i, j, 0] = 255 - image_negasi[i, j, 0]
                    image_negasi[i, j, 1] = 255 - image_negasi[i, j, 1]
                    image_negasi[i, j, 2] = 255 - image_negasi[i, j, 2]
        image_negasi = Image.fromarray(image_negasi)
        image_negasi_resize = image_negasi.resize((197, 197), Image.Resampling.LANCZOS)
        image_negasi = ImageTk.PhotoImage(image_negasi_resize)
        label_image_negasi = tk.Label(window, image=image_negasi)
        label_image_negasi.image = image_negasi
        label_image_negasi.grid(row=2, column=1, pady=(0,20), padx=(20, 0), sticky="w")

        #Image Threshold
        image_threshold = skimage.io.imread(fname=image_name)
        threshold = 150
        if image_depth == 8:
            for i in range(0, image_resolution_y):
                for j in range(0, image_resolution_x):
                    if image_threshold[i, j] < threshold:
                        image_threshold[i, j] = 0
                    else:
                        image_threshold[i, j] = 255
        elif image_depth == 24:
            for i in range(0, image_resolution_y):
                for j in range(0, image_resolution_x):
                    if image_threshold[i, j, 0] < threshold:
                        image_threshold[i, j, 0] = 0
                    else:
                        image_threshold[i, j, 0] = 255
                    if image_threshold[i, j, 1] < threshold:
                        image_threshold[i, j, 1] = 0
                    else:
                        image_threshold[i, j, 1] = 255
                    if image_threshold[i, j, 2] < threshold:
                        image_threshold[i, j, 2] = 0
                    else:
                        image_threshold[i, j, 2] = 255
        image_threshold = Image.fromarray(image_threshold)
        image_threshold_resize = image_threshold.resize((197, 197), Image.Resampling.LANCZOS)
        image_threshold = ImageTk.PhotoImage(image_threshold_resize)
        label_image_threshold = tk.Label(window, image=image_threshold)
        label_image_threshold.image = image_threshold
        label_image_threshold.grid(row=2, column=2, pady=(0,20), padx=(20, 0), sticky="w")

window.mainloop()
```
